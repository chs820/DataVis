```python
import matplotlib.pyplot as plt
import numpy as np

%matplotlib inline
```

```python
plt.figure()

languages = ['Python','SQL','Java','C++','JavaScript']
pos = np.arange(len(languages))
popularity = [56, 39, 34, 34, 29]

plt.bar(pos, popularity, align='center')
plt.xticks(pos, languages)
plt.ylabel('% Popularity')
plt.title('Top 5 Languages for Math & Data \nby % popularity on Stack Overflow', alpha=0.8)

plt.show()
```

```python
fontsize=20
alpha=0.8

ax = plt.figure(figsize=(12,9))

languages = ['Python','SQL','Java','C++','JavaScript']
pos = np.arange(len(languages))
popularity = [56, 39, 34, 34, 29]

bars = plt.bar(pos, popularity, align='center', linewidth=0, color='lightslategrey')
bars[0].set_color('#1F77B4')

plt.xticks(pos, languages, alpha=alpha, fontsize=fontsize)
#plt.ylabel('% Popularity', alpha=0.8)
plt.title('Top 5 Languages for Math & Data \nby % popularity on Stack Overflow', alpha=alpha, fontsize=fontsize)

plt.tick_params(
    top='off',
    bottom='off',
    left='off',
    right='off',
    labelleft='off',
    labelbottom='on'
    )

for spine in plt.gca().spines.values():
    spine.set_visible(False)

for bar in bars:
    plt.gca().text(bar.get_x() + bar.get_width()/2, bar.get_height() - 4, 
                   str(int(bar.get_height())) + '%', ha='center', color='w', alpha=alpha, fontsize=fontsize)

plt.show();
```
